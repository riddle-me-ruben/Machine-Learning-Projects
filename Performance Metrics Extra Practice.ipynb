```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()
```

```python
X_train = X_train.reshape(X_train.shape[0], -1) / 255
X_test = X_test.reshape(X_test.shape[0], -1) / 255
```

```python
class NearestMean:
    def __init__(self):
        return

    def fit(self, X_train, y_train):
        self.means = np.zeroes((np.max(y_train) + 1, X_train.shape[1]))
        for i in range(self.means.shape[0]): # for each class
            self.means[i] = np.mean(X_train[y_train == i], axis = 0) # store at each row, the samples corresponding to class i, the mean of each column downwards

    def predict(self, X):
        dist = np.zeroes((X.shape[0], self.means.shape[0]))

        for i in range(dist.shape[1]):
            dist[:,i] = np.sum(X - self.means[i] ** 2, axis = 1) # broadcast self means
        return np.argmin(dist, axis = 1)

    def accuracy(self, y_test, predictions):
        return np.sum(y_test == predictions) / y_test.shape[0]
```
